```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dates_health = pd.date_range(start="2023-01-01", periods=24, freq="MS")
admissions = (np.sin(np.linspace(0, 3 * np.pi, 24)) * 50 + 300 + np.random.normal(0, 15, 24)).astype(int)
healthcare_df = pd.DataFrame({"Month": dates_health, "Patient_Admissions": admissions})
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(x=healthcare_df['Month'],y=healthcare_df['Patient_Admissions'],data=healthcare_df,linewidth=2,linestyle='--',marker='o')
plt.ylabel("No.of Admissions")
plt.xlabel("Month")
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

quarters = pd.date_range(start="2023-01-01", periods=8, freq="QS")
solar = np.cumsum(np.random.uniform(20, 40, len(quarters)))
wind = np.cumsum(np.random.uniform(10, 30, len(quarters)))
fossil = np.cumsum(np.random.uniform(50, 80, len(quarters)))
energy_df = pd.DataFrame({
    "Quarter": quarters,
    "Solar_MW": solar,
    "Wind_MW": wind,
    "Fossil_MW": fossil
})
energy_df

```

```python
plt.figure(figsize=(10, 6))
plt.stackplot(
    energy_df["Quarter"],
    energy_df["Solar_MW"],
    energy_df["Wind_MW"],
    energy_df["Fossil_MW"],
    labels=["Solar", "Wind", "Fossil Fuel"],
    alpha=0.8
)

plt.legend(loc="upper left")
plt.title("Quarterly Energy Generation by Source")
plt.xlabel("Quarter")
plt.ylabel("Cumulative Generation (MW)")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
months_trans = pd.date_range(start="2024-01-01", periods=12, freq="MS")
ridership = (np.linspace(20000, 30000, 12) + np.random.normal(0, 1000, 12)).astype(int)
delays = np.random.uniform(5, 20, 12)
transport_df = pd.DataFrame({
    "Month": months_trans,
    "Ridership": ridership,
    "Avg_Delay_Minutes": delays
})
fig, ax1 = plt.subplots(figsize=(10,6))
ax1.bar(transport_df["Month"], transport_df["Ridership"],
        color="skyblue", label="Ridership")
ax1.set_ylabel("Ridership", color="blue")
ax1.tick_params(axis='y', labelcolor="blue")
ax2 = ax1.twinx()
ax2.plot(transport_df["Month"], transport_df["Avg_Delay_Minutes"],
         color="red", marker="o", linewidth=2, label="Avg Delay (min)")
ax2.set_ylabel("Average Delay (Minutes)", color="red")
ax2.tick_params(axis='y', labelcolor="red")
plt.title("Monthly Ridership vs Average Delays (2024)")
fig.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
num_plants = 100
production_volume = np.random.uniform(1000, 10000, num_plants)
defect_rate = np.clip(0.02 + 0.00001 * production_volume + np.random.normal(0, 0.005, num_plants), 0, 0.1)
manufacturing_df = pd.DataFrame({
    "Production_Volume": production_volume.astype(int),
    "Defect_Rate": defect_rate
})
manufacturing_df
plt.scatter(manufacturing_df["Production_Volume"],manufacturing_df["Defect_Rate"])
plt.xlabel("Production_Volume")
plt.ylabel("Defect_rate")
plt.title("Relationship production_volume and defect_rate")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dates_fin = pd.date_range(start="2025-01-01", periods=60, freq="B")
stock_a = np.cumsum(np.random.normal(0.1, 1, len(dates_fin))) + 100
stock_b = np.cumsum(np.random.normal(0.05, 1.2, len(dates_fin))) + 80
finance_df = pd.DataFrame({
    "Date": dates_fin,
    "Stock_A_Close": stock_a,
    "Stock_B_Close": stock_b
})
plt.figure(figsize=(10,6))
plt.plot(finance_df["Date"],finance_df["Stock_A_Close"],label="Stock_A_Close")
plt.plot(finance_df["Date"],finance_df["Stock_B_Close"],label="Stock_B_Close")
plt.xlabel("Date")
plt.ylabel("Stocks")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
satisfaction_scores = np.clip(np.random.normal(4.0, 0.7, 300), 1, 5)
satisfaction_df = pd.DataFrame({"Score": satisfaction_scores})
def categorize(score):
    if score >= 4.5:
        return "Promoter"
    elif score >= 3.0:
        return "Passive"
    else:
        return "Detractor"
satisfaction_df["Category"] = satisfaction_df["Score"].apply(categorize)
category_counts = satisfaction_df["Category"].value_counts().reset_index()
category_counts.columns = ["Category", "Count"]



```

```python

print(category_counts)

fig, axes = plt.subplots(1, 2, figsize=(12,5))
axes[0].hist(satisfaction_df["Score"], bins=10, color="skyblue", edgecolor="black")
axes[0].set_title("Distribution of Guest Satisfaction Scores")
axes[0].set_xlabel("Score (1–5)")
axes[0].set_ylabel("Number of Guests")

axes[1].pie(
    category_counts["Count"],
    labels=category_counts["Category"],
    autopct="%1.1f%%",
    startangle=90,
    colors=["green", "gold", "red"]
)
axes[1].set_title("Category Breakdown")
plt.tight_layout()
plt.show()
```
