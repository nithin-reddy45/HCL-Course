```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Departments=['HR','IT','Finance','Sales','Marketing']
Employees=[20,40,15,30,20]
Budget=[12,18,10,20,15]
salaries={
    'HR':[30,35,32,40,45],
    'IT':[60,65,70,75,80],
    'Finance':[50,55,53,58,60],
    'Sales':[45,48,50,52,55],
    'Marketing':[40,42,45,47,50]
}
plt.figure(figsize=(8,5))
plt.barh(Departments,Employees,color='orange')
plt.xlabel('Departments')
plt.ylabel('No.of Employees')
plt.title('horizontal bar chart for number of employees per department.')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(Departments,Employees,color="blue",label="Employees")
plt.bar(Departments,Budget,bottom=Employees,label="Budget(lakhs)",color="green")
plt.title("Employees & Budget Allocation per Department")
plt.ylabel("Count / Allocation")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.boxplot(salaries.values(),labels=salaries.keys())
plt.title("Salary Distribution per Department (Boxplot)")
plt.xlabel("Departments")
plt.ylabel("Salary (in K)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.violinplot(salaries.values(), showmeans=True, showmedians=True)
plt.xticks(range(1, len(Departments) + 1), Departments)
plt.title("Salary Distribution per Department (Violin Plot)")
plt.xlabel("Departments")
plt.ylabel("Salary (in K)")
plt.show()
```

For HR analytics, boxplots and violin plots are most useful as they show salary distribution, variations, and outliers. They help identify pay gaps and fairness across departments. The horizontal bar chart is good for comparing employee strength, while the stacked bar chart shows both staff and budget together. Overall, bar charts give quick comparisons, but boxplots and violin plots provide deeper insights for HR decisions


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
months=["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales_2023 = [200, 220, 250, 270, 300, 320]
sales_2024 = [210, 230, 260, 280, 310, 340]
Orders = [20, 30, 25, 40, 35, 45]
Returns = [2, 3, 1, 4, 2, 3]
plt.figure(figsize=(8, 5))
plt.fill_between(months, sales_2023, alpha=0.5, label="Sales 2023")
plt.fill_between(months, sales_2024, alpha=0.5, label="Sales 2024")
plt.title("Sales Trend Comparison (2023 vs 2024)")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
x = np.arange(len(months))
width = 0.35
plt.figure(figsize=(8, 5))
plt.bar(x - width/2, sales_2023, width, label="Sales 2023")
plt.bar(x + width/2, sales_2024, width, label="Sales 2024")
plt.xticks(x, months)
plt.title("Monthly Sales Comparison (2023 vs 2024)")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(months, Orders, s=[r*100 for r in Returns], alpha=0.6, color="skyblue", edgecolors="black")
plt.title("Orders vs Returns (Bubble Size = Returns)")
plt.xlabel("Months")
plt.ylabel("Orders")
plt.show()
```

The area chart shows that 2024 sales are consistently higher than 2023, reflecting steady growth.
The side-by-side bar chart highlights monthly differences, making it easy to compare performance.
The bubble chart links orders with returns, showing the balance between demand and quality.
April has the largest bubble, indicating the highest returns despite good sales volume.
These insights help managers track growth, evaluate performance, and address product issues.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
days=['Mondat','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
visits=[120,150,180,170,200,240,190]
plt.plot(days,visits,linestyle='--',color='red',marker='o',linewidth=2.5,markersize=8)
plt.title("Website Traffic Trend (Campaign Week)",fontsize=16)
plt.xlabel("Day of Week",fontsize=12)
plt.ylabel("Number of Visits",fontsize=12)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
revenue = [80, 120, 160, 190, 220]
plt.figure(figsize=(8,5))
plt.fill_between(months, revenue, color="lightgreen", alpha=0.7)
plt.plot(months, revenue, color="darkgreen", linewidth=2.5)
plt.title("Monthly Revenue Accumulation")
plt.xlabel("Months")
plt.ylabel("Revenue (in thousands)")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = {
    "workout_type": ["Gym", "Yoga", "Running", "Cycling", "Swimming","Gym", "Gym", "Running", "Cycling", "Jogging"],
    "gender": ["Male", "Female", "Male", "Female", "Male","Female", "Male", "Female", "Male", "Female"],
    "satisfaction": [8, 7, 6, 9, 8, 7, 8, 5, 6, 9]
}
df = pd.DataFrame(data)
df
```

```python

plt.figure(figsize=(7,4))
sns.countplot(x="workout_type", data=df, palette="pastel")
plt.title("Workout Popularity")
plt.ylabel("Count")
plt.show()
```

```python

plt.figure(figsize=(7,4))
sns.countplot(x="workout_type", hue="gender", data=df, palette="Set2")
plt.title("Workout Popularity by Gender")
plt.ylabel("Count")
plt.show()
```

```python

plt.figure(figsize=(7,4))
sns.barplot(x="workout_type", y="satisfaction", data=df, estimator="mean", palette="pastel")
plt.title("Average Satisfaction per Workout")
plt.ylabel("Mean Satisfaction")
plt.show()
```

```python
sns.boxplot(x="workout_type", y="satisfaction", data=df, palette="Set3")
plt.title("Satisfaction Distribution by Workout")
plt.ylabel("Satisfaction")
plt.show()
```

```python

plt.figure(figsize=(7,4))
sns.lineplot(x="workout_type", y="satisfaction", hue="gender", data=df,  marker="o")
plt.title("Satisfaction Trends by Gender")
plt.ylabel("Mean Satisfaction")
plt.show()
```

```python


plt.figure(figsize=(7,4))
sns.barplot(x="workout_type", y="satisfaction", hue="gender", data=df, estimator="mean", palette="Set1")
plt.title("Average Satisfaction by Workout & Gender")
plt.ylabel("Mean Satisfaction")
plt.show()
```
